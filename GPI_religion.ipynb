```python
import pandas as pd
import matplotlib.pyplot as plt
import colorsys
plt.style.use('seaborn-talk')
%matplotlib inline
```

```python
#We will explore this dataframe in order to find any relation between the index presented
#GPD - Gross domestic product (GDP) is the total monetary or market value of all the finished 
#goods and services produced within a countryâ€™s borders in a specific time period.
#GPI - s a report produced by the Institute for Economics & Peace (IEP) which measures the 
#relative position of nations' and regions' peacefulness. The score is measure between 1-5,
#the closest to 1 the better

```

```python
#To have a general idea of our dataframe
df.describe()
```

```python
#top 10 non religious country
df2 = df.nlargest(10, 'percentage_non_religious')
df2
```

```python
#The first conclusion we can draw from this is that  9 from the top 10 non religious contry have a 
#gpi lower than the average
df3 = df2[['country', 'percentage_non_religious']]
df3.plot.bar(x='country', y='percentage_non_religious')
df3
```

```python
#Lets see the top 10 countries based on GDP
df4 = df.nlargest(10, 'gdp_pc')
df4.plot.bar(x='country', y='gdp_pc')
plt.show()
```

```python
#we can see that singapore has the highest GDP, lets take a look at Singapore information
dfsing=df4[df4['country'] == 'Singapore'].index[0]
```
